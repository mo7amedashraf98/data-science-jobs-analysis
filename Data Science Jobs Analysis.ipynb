## Data Visualization 

Answering questions using visualization.

### Q1: What's the percentage of Data Scientists through the years? 

#### 1- Distribution by pie chart

```python
def work_year(df,work_year): 
    plt.figure(figsize=[10,10]);
    df[work_year].value_counts().plot(kind='pie',label='work_year',autopct='%0.2f')
    plt.legend();
    plt.title('Percentage of Data Scientists by year');
    plt.ylabel('Work Year');
work_year(df,'work_year')
```

#### 2- Distribution by bar chart

```python
plt.figure(figsize=(10,10))
snb.countplot(data=df, x='work_year')
plt.title('Number of Data Scientists each year')
plt.show()
```

```python
df['work_year'].value_counts()
```

### Q2: What's the distribution by experience levels? 

#### 1- Pie chart

```python
def experience_level(df,experience_level):
    plt.figure(figsize=[10,10])
    df[experience_level].value_counts().plot(kind="pie", label="experience_level", autopct='%0.2f');
    plt.title('Experience level distribution');
    plt.legend();
    plt.ylabel('experience_level')
experience_level(df,'experience_level')
```

#### 2- Bar chart

```python
plt.figure(figsize=[10,10])
snb.countplot(data=df, x='experience_level')
plt.title('Experience level distribution')
plt.show()
```

```python
df['experience_level'].value_counts()
```

### Q3: What's the distribution by employment type?

### Q4: What's the distribution by job title? 

#### 1- Pie chart

```python
def job_title(df,job_title):
    plt.figure(figsize=[10,10])
    df['job_title'].value_counts().head(15).plot(kind='pie', label='job_title',autopct='%0.2f',fontsize=10)
    plt.title('Distribution by job title (top 15)');
    plt.ylabel('job_title');
job_title(df,job_title)
```

#### 2- Bar chart: 

```python
plt.figure(figsize=[14,14])
snb.countplot(data=df,x='job_title')
plt.xticks(rotation=90)
plt.title('Distribution by job title')
plt.show()
```

```python
df['job_title'].value_counts().head(15)
```

### Q8: What's the distribution of the experience level relative to the work year? 

```python
df.pivot_table(index='experience_level', columns='work_year', values='employment_type',aggfunc=['count','mean']).plot(kind='pie',subplots=True, autopct='%0.2f', figsize=(20,10), title='Distribution of the level of experience & type of employment relative to the year of work', fontsize=15)
plt.show()
```

```python
df.pivot_table(index='experience_level',columns='work_year',values='employment_type', aggfunc=['count']).plot(kind='bar', figsize=[10,10])
plt.show()
```

```python
df.pivot_table(index='experience_level', columns='work_year', values='employment_type', aggfunc=['count'])
```

### Q9: What's the distribution of the experience level relative to company size? 

```python
df.pivot_table(index='experience_level', columns='company_size', values='employment_type', aggfunc=['count','mean']).plot(kind='pie',subplots=True, autopct='%0.2f', figsize=(20,10),title='Distribution of the employment type & company size relative to the experience level', fontsize=15)
plt.show()
```

### Q10: What's the average salary in USD relative to the level of experience? 

```python
df.groupby('experience_level')['salary_in_usd'].agg(['mean','median']).plot(kind='line', figsize=[10,10], title=('Distribution of the average salary in USD relative to the experience level'))
plt.show()
```

```python
df.groupby('experience_level')['salary_in_usd'].agg(['mean','median']).sort_values(by='mean', ascending=False)
```

### Q11: What's the average salary in USD relative to work year? 

```python
df.groupby('work_year')['salary_in_usd'].agg(['mean']).plot(kind='bar', figsize=[10,10], title=('Distribution of the average salary in USD relative to work year'))
plt.show()
```

```python
df.groupby('work_year')['salary_in_usd'].agg(['mean']).sort_values(by='mean', ascending=False)
```
